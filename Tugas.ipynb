4.visualisasikan jumlah nilai unik

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['released_year'], bins=10, kde=True)
plt.title('Histogram tahun liris')
plt.xlabel('released_year')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

8. Filter data

```python

# Filter berdasarkan tahun
filtered_by_year = data[data['released_year'] == 2020]
print("\nFilter berdasarkan released_year == 2020:")
print(filtered_by_year)
```

```python

# Filter berdasarkan bulan
filtered_by_month_range = data[(data['released_month'] >= 1) & (data['released_month'] <= 6)]
print("\nFilter berdasarkan released_month antara 1 dan 6:")
print(filtered_by_month_range)
```

9. Box plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[['artist_count','in_spotify_playlists', 'bpm', 'danceability_%']])
plt.title('Box Plot Metrik Musik')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

10. Korelasi

```python

# Mengatur ukuran figure
plt.figure(figsize=(12, 6))

# Menghitung korelasi dan membuat heatmap
corr_matrix = data.select_dtypes('number').corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
